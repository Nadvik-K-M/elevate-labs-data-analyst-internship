```python
#null values
missing_values = slots.isnull().sum()
print("Missing values ",missing_values)

col_missing_values = slots["HH Impressions"].isnull().sum()
print(" \n Missing values for HH Impressions ",col_missing_values)

```

```python
#Percentage wise Null values
NetworkType = slots['Network Type'].isnull().sum()/slots.shape[0]*100
NetworkFeed = slots['Network Feed'].isnull().sum()/slots.shape[0]*100
HHImpressions = slots['HH Impressions'].isnull().sum()/slots.shape[0]*100
HHRatings = slots['HH Ratings'].isnull().sum()/slots.shape[0]*100
print(f"Null values Percentage are:  \n 1 NetworkType = {NetworkType} \n 2 NetworkFeed = {NetworkFeed} \n 3 HHImpressions = {HHImpressions}\n 4 HHRatings  {HHRatings} ")
```

```python
slots_cleaned = slots.drop_duplicates()
slots_cleaned
```

```python
# Filling Null Values with mean and Mode
air_time = slots['Air Time'].mode()
slots['Air Time'].fillna(air_time[0],inplace = True)

network_type = slots['Network Type'].mode()
slots['Network Type'].fillna(network_type[0],inplace = True)

Network_DMA = slots['Network DMA'].mode()
slots['Network DMA'].fillna(Network_DMA[0],inplace = True)

Network_Feed = slots['Network Feed'].mode()
slots['Network Feed'].fillna(Network_Feed[0],inplace = True)
```

```python
HH_Impressions = slots['HH Impressions'].mean()
slots['HH Impressions'].fillna(HH_Impressions,inplace = True)

HH_Ratings  = slots['HH Ratings'].mean()
slots['HH Ratings'].fillna(HH_Ratings,inplace = True)

Spot_Cost  = slots['Spot Cost'].mean()
slots['Spot Cost'].fillna(Spot_Cost,inplace = True)
```
